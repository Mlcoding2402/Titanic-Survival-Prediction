```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
print(df.head())
print(df.info())
print(df.isnull().sum())
```

```python
#missing values handle

df["Age"].fillna(df["Age"].median(),inplace=True)
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
```

```python
df.drop(["Passenger","Name","Ticket"],axis=1,inplace=True,errors="ignore")
df.drop("Cabin",axis=1,inplace=True)
print(df)
```

```python
#Encoding (convert categorical data into numeric data)

le = LabelEncoder()
df["Sex"] = le.fit_transform(df["Sex"])
df["Embarked"] = le.fit_transform(df["Embarked"])
```

```python
#features and Target
print(df)
X = df.drop("Survived",axis = 1)
Y = df["Survived"]
```

```python
#Train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state =42)
```

```python
#Train model using logisticregression
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train,Y_train)
```

```python
#prediction
Y_pred = lr_model.predict(X_test)
#Accuracy
accuracy = accuracy_score(Y_test,Y_pred)
print("Accuracy:",accuracy)
#confussion Matrix
print(confusion_matrix(Y_test,Y_pred))
```

```python
#train model using decision tree
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train,Y_train)
```

```python
#prediction
Y_pred = dt_model.predict(X_test)
#Accuracy
accuracy = accuracy_score(Y_test,Y_pred)
print("Accuracy:",accuracy)
#confussion Matrix
print(confusion_matrix(Y_test,Y_pred))
```

```python
#train model using random forest
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train,Y_train)
```

```python
#prediction
Y_pred = rf_model.predict(X_test)
#Accuracy
accuracy = accuracy_score(Y_test,Y_pred)
print("Accuracy:",accuracy)
#confussion Matrix
print(confusion_matrix(Y_test,Y_pred))
```

```python
#train model using K-Nearest Neighbors(KNN)
from sklearn.neighbors import KNeighborsClassifier
kn_model = KNeighborsClassifier()
kn_model.fit(X_train,Y_train)
```

```python
#prediction
Y_pred = kn_model.predict(X_test)
#Accuracy
accuracy = accuracy_score(Y_test,Y_pred)
print("Accuracy:",accuracy)
#confussion Matrix
print(confusion_matrix(Y_test,Y_pred))
```

```python
#train model using Support vector machine
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,Y_train)
```

```python
#prediction
Y_pred = model.predict(X_test)
```

```python
#Accuracy
accuracy = accuracy_score(Y_test,Y_pred)
print("Accuracy:",accuracy)
```

```python
#confussion Matrix
print(confusion_matrix(Y_test,Y_pred))
```
